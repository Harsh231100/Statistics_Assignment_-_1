### ○	Co-efficient of Range

### ○	Co-efficient of Standard Deviation
### ○	Co-efficient of Quartile Deviation
### ○	Co-efficient of Mean Deviation


```python
import pandas as pd, numpy as np
```

```python
class data():
    def __init__(self,data):
        self.data=data
    
    def arithmetic_mean(self):
        return sum(self.data) / len(self.data)

    def geometric_mean(self):
        product = 1
        n = len(self.data)
        for num in self.data:
            product *= num
        return product**(1/n)

    def harmonic_mean(self):
        denominator = 0
        for x in self.data:
            denominator += 1/x
        return len(self.data) / denominator

    def mode(self):
        counts = {}
        for num in self.data:
            counts[num] = counts.get(num, 0) + 1
        mode = [key for key, value in counts.items() if value == max(counts.values())]
        return mode
    
    def median(self):
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        if n%2==0:
            return(sorted_data[n//2]+sorted_data[n//2-1])/2
        else:
            median=sorted_data[n//2]
        return median

    def describe_data(self):
        n = len(self.data)
        mean = sum(self.data) / n
        variance = sum((x - mean) ** 2 for x in self.data) / (n - 1)
        std_deviation = variance ** 0.5
        return n, mean, variance, std_deviation, min(self.data), max(self.data)

    def variance(self):
        n = len(self.data)
        mean = sum(self.data) / n
        variance = sum((x - mean) ** 2 for x in self.data) / (n - 1)
        return variance

    def standard_deviation(self):
        return self.variance() ** 0.5

    def maximum(self):
        return max(self.data)

    def minimum(self):
        return min(self.data)

    def calculate_range(self):
        return self.maximum() - self.minimum()

    def iqr(self):
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        q1 = sorted_data[n // 4]
        q3 = sorted_data[(3 * n) // 4]
        return q3 - q1

    def quartiles(self):
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        q1 = sorted_data[n // 4]
        q2 = sorted_data[n // 2]
        q3 = sorted_data[(3 * n) // 4]
        return q1, q2, q3

    def coefficient_of_range(self):
        return (self.maximum() - self.minimum()) / (self.maximum() + self.minimum())

    def coefficient_of_variation(self):
        return self.standard_deviation() / self.arithmetic_mean()

    def coefficient_of_standard_deviation(self):
        return self.standard_deviation() / self.arithmetic_mean()

    def coefficient_of_quartile_deviation(self):
        q1, q3 = self.quartiles()[0], self.quartiles()[2]
        return (q3 - q1) / (q3 + q1)

    def coefficient_of_mean_deviation(self):
        mean = self.arithmetic_mean()
        mean_deviation = [abs(x - mean) for x in self.data]
        return sum(mean_deviation) / len(mean_deviation) if len(mean_deviation) != 0 else None
```
